```python
# setup strings for status, categories and origin
socialStyle = {115:"Employer",116:"VAT Payer",117:"CRAM-independent",118:"AKM-independent",120:"Co-working spouse",130:"Employees indistinguishably",131:"Top executive",132:"Employees at the highest level",133:"Employees intermediate level",134:"Employees at basic level",135:"Other employees",136:"Employees end of November (in addition to the primary or secondary job)",138:"Employees at end of November",139:"Secondary job",140:"Employed earner without national work (not resident in DK)",141:"Employed executives (not resident in DK)",142:"Employed workers at the highest level (not resident in DK)",143:"Employed workers intermediate level (not resident in DK)",144:"Employed workers at basic level (not resident in DK)",145:"Employed other employees (not resident in DK)",150:"Jobs for employees without national work (not resident in DK)",151:"Jobs for senior executives (not resident in DK)",152:"Jobs for employees at the highest level (not resident in DK)",153:"Jobs for employees intermediate level (not resident in DK)",154:"Jobs for employees at basic level (not resident in DK)",155:"Jobs for medarbejdere the basic level (not resident in DK)",200:"Netto unemployed",201:"Brutto unemployed",310:"learners",315:"Flex benefit",316:"Partial unemployment",317:"Employed without pay",318:"Leave of absence from unemployment",319:"Training measures/counseling and skills upgrading",320:"Special/work placements",321:"undisclosed work placements",322:"Maternity benefits",323:"Sickness benefits",324:"Early retirement benefit",325:"Transitional benefit",326:"Cash benefits",327:"Rehabilitation benefit",328:"Civil servant pension",329:"Pension",330:"Others outside the labor force",331:"Disability pensioners",332:"Introductory benefit",333:"Integration education",334:"Unemployment allowance",335:"Activation acc. social security statistics register",400:"Child or young (that is under 16 years)"}
socialGroup = {1: "Employed", 2: "Unemployed", 3: "Outside workforce"}
category = {1: "Under 16 years", 2: "16-24 years", 3: "25-29 years", 4: "30-59 years", 5: "60-66 years", 6: "67 years and over"};
origin = {1: "Danish", 2: "immigrant", 3: "Progeny"}
district = {1:"Indre By", 2:"Østerbro", 3:"Nørrebro", 4:"Vesterbro/Kgs. Enghave", 5:"Valby", 6:"Vanløse", 7:"Brønshøj-Husum", 8:"Bispebjerg", 9:"Amager Øst", 10:"Amager Vest", 99: "Outside classification"}

# import csv utilities
import csv

# load data into dict
socStatus = []
with open('Socio_Stat_Alder_Herkomst1.csv') as f:
    reader = csv.DictReader(f, delimiter=',')
    for row in reader:
        if int(row['AAR']) in range(2003, 2014):
            socStatus.append({'year': row['AAR'], 'district': row['BYDEL'], 'age': row['ALDERSKAT'], 'origin': row['IETYPE'], 'socialGroup': row['SOC_GRUPPE'], 'socialStyle': row['SOCSTIL'], 'persons': row['PERSONER']})
```

```python
# Make plot inline
%matplotlib inline
#import matplotlib, numpy, itemgetter and counter
import matplotlib.pyplot as plt
import numpy as np
from operator import itemgetter
from collections import Counter

# set figure size
fig = plt.figure(figsize=(15, 25))

# set standard variables
width = 0.7
plotNumber = 1

# plot the crimes per year by using a forloop
for dist in district.keys():  
    plt.subplot(7, 2, plotNumber)
    data = Counter()
    for status in socStatus:
        if int(status['district']) == dist and int(status['socialGroup']) == 2:
            if data.has_key(status['year']):
                data[status['year']] += int(status['persons'])
            else:
                data[status['year']] = int(status['persons'])
    
    data = sorted(data.items(), key=itemgetter(0))
    labels = [item[0] for item in data]
    values = [item[1] for item in data]
    indexes = np.arange(len(labels))
    plt.bar(indexes, values, width)
    plt.xticks(indexes + width * 0.5, labels)
    plt.ylabel('Unemployed People')
    title = unicode(district[dist], "utf-8")
    plt.title(title)
    plotNumber += 1

plt.show()
```
